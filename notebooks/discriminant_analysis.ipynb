# Discriminant Analysis with REGAIN

```python
import numpy as np
from regain.datasets import make_dataset
```

```python
from sklearn.metrics import accuracy_score
```

```python
data = make_dataset(n_samples=500, mode='sin')
```

```python
c_data = np.vstack(data.data)
```

```python
n_time = data.data.shape[0]
```

```python
y = np.repeat(np.arange(n_time), [X.shape[0] for X in data.data])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    c_data, y, test_size=0.1, random_state=42)
```

```python
from regain import discriminant_analysis; reload(discriminant_analysis);

da = discriminant_analysis.DiscriminantAnalysis(beta=0, alpha=1.5).fit(X_train, y_train)
```

```python
y_pred = da.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

qda = QuadraticDiscriminantAnalysis(store_covariance=True, reg_param=0.5).fit(X_train, y_train)
```

```python
y_pred = qda.predict(X_test)
accuracy_score(y_test, y_pred)
```
